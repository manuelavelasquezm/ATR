```python

fecha_inicio=date.today()-datetime.timedelta(days=100)

fecha_final=date.today()

```

```python
def exportarATR(acciones, n):


  from datetime import date, timedelta

  fecha_inicio = date.today() - timedelta(100)

  fecha_final =  date.today()

  def calcularTR(acciones):
    TR = []

    for accion in acciones['Low'].columns:

      aux = []

      for i in acciones.index:

        if i > 0:
          aux.append(max(acciones['High'][accion][i]- acciones['Low'][accion][i], 
                        abs(acciones['High'][accion][i] - acciones['Adj Close'][accion][i -1]),
                        abs(acciones['Low'][accion][i] - acciones['Adj Close'][accion][i -1])))
        else:
          aux.append(acciones['High'][accion][i]- acciones['Low'][accion][i])

      TR.append(aux)

    return(TR)
  
  def calcularATR(TR, n):
    ATR = []
    for accion in TR.columns:
      aux = []
      for i in range(len(TR)):
        if i > n-1:
          aux.append((aux[-1]*(n - 1) + TR[accion][i])/n)

        if i == 13:
          aux.append(TR[accion][0:n].mean())
          
      ATR.append(aux)

    return(ATR)

  datos = yf.download(acciones, start = fecha_inicio, end = fecha_final, progress = False)

  indices = datos.index

  datos.index = range(len(datos))
  TR = pd.DataFrame(np.array(calcularTR(datos)).T, columns = datos['Low'].columns)
  TR.index = indices

  ATR = pd.DataFrame(np.array(calcularATR(TR, 14)).T, columns = datos['Low'].columns)
  ATR.index = TR.index[-len(ATR):]

  atrs = pd.DataFrame(index = datos['Low'].columns, columns = ['Fecha', 'Cierre', 'ATR', 'ATR/2', 'S1', 'S2', 'R1', 'R2'])

  atrs['Fecha'] = pd.to_datetime(ATR.index[-1])
  atrs['Fecha'] = atrs['Fecha'].dt.strftime('%d/%m/%Y')
  atrs['Cierre'] = datos['Adj Close'].iloc[-1]
  atrs['ATR'] = ATR.iloc[-1]
  atrs['ATR/2'] = ATR.iloc[-1]/2
  atrs['S1'] = atrs['Cierre'] - atrs['ATR/2']
  atrs['S2'] = atrs['Cierre'] - atrs['ATR']
  atrs['R1'] = atrs['Cierre'] + atrs['ATR/2']
  atrs['R2'] = atrs['Cierre'] + atrs['ATR']

  atrs.to_excel('ATR_Acciones.xlsx', sheet_name = 'ATR')

  return atrs
```
